# KNN Iris Classifier 


KNN used for classifier
Compares to most similar data points 







### Import Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Import & Explore Data

```python
iris = pd.read_csv('../datasets/iris/iris.csv')
```

```python
iris.head()
```

```python
iris.tail()
```

```python
sns.pairplot(iris, hue = 'Species', vars = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm' ])
```

```python
sns.scatterplot(x = 'SepalLengthCm', y = 'PetalLengthCm', hue = 'Species', data = iris)
```

```python
# plot corrilations
plt.figure(figsize =(30,20))

sns.heatmap(iris.corr(), annot = True)
```

### Data Cleaning & Prep

```python
X = iris.drop(['Species'], axis = 1)
```

```python
X.shape
```

```python
X.head()
```

```python
y = iris['Species']
```

```python
y.shape
```

```python
y
```

```python
# transform y data into digits (0,1)
from sklearn.preprocessing import LabelEncoder
labelencoder_y = LabelEncoder()
y = labelencoder_y.fit_transform(y)
```

```python
y
```

```python
# Create train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20, random_state = 5, stratify=y)
```

```python
X_train.shape
```

```python
X_test.shape
```

### Train & test Model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier = KNeighborsClassifier(n_neighbors=3, metric = 'minkowski', p=2)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
print(y_pred)
print(y_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(cm,annot = True)
```

```python
print(classification_report(y_test, y_pred))
```
